```python
# Importing the libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Reading and defining the dataset

df = pd.read_csv("Mealapp_dataset.csv", encoding='Latin-1')
```

```python
df.head()
```

```python
df = df[["Meal_description", "Meal_category", "Meal_name"]]
```

```python
df.shape
```

```python
df = df.sample(n=10000, replace=False, random_state=33).reset_index(drop=True, )
```

```python
df.shape
```

```python
df.head()
```

```python
df_ = df.drop_duplicates(["Meal_name"]).reset_index(drop=True)

df_.shape
```

```python
df_
```

```python
des = df_["Meal_description"]
cat = df_["Meal_category"]
meal = df_["Meal_name"]
data = []

for i in range(len(des)):
    str_ = "{} {} {}".format(des[i], cat[i], meal[i])
    data.append(str_.lower())
```

```python
vec = TfidfVectorizer()
vec_ = vec.fit_transform(data)
```

```python
sim = cosine_similarity(vec_)
```

```python
df1 = pd.DataFrame(sim, columns=df_['Meal_name'], index=df_['Meal_name']).reset_index()
df1.head()
```

```python
df1.to_csv("processed.csv", index=False)
```
